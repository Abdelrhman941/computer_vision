> ## Changing the image lighting color >>

> 

```python
# Define the constant value to add
C = 50

# Create reddish image
img_reddish = img.copy()
img_reddish[:, :, 2] = cv2.add(img_reddish[:, :, 2], C)

# Create greenish image
img_greenish = img.copy()
img_greenish[:, :, 1] = cv2.add(img_greenish[:, :, 1], C)

# Create bluish image
img_bluish = img.copy()
img_bluish[:, :, 0] = cv2.add(img_bluish[:, :, 0], C)

# Create yellowish image (R + G)
img_yellowish = img.copy()
img_yellowish[:, :, 1] = cv2.add(img_yellowish[:, :, 1], C)     # G
img_yellowish[:, :, 2] = cv2.add(img_yellowish[:, :, 2], C)     # R

# Display the images
cv2.imshow('Original Image', img)
cv2.imshow('Reddish Image', img_reddish)
cv2.imshow('Greenish Image', img_greenish)
cv2.imshow('Bluish Image', img_bluish)
cv2.imshow('Yellowish Image', img_yellowish)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# the code by matplotlib >>
import matplotlib.pyplot as plt

# Create a figure and a set of subplots
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Original Image
axs[0, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0, 0].set_title('Original Image')
axs[0, 0].axis('off')

# Reddish Image
axs[0, 1].imshow(cv2.cvtColor(img_reddish, cv2.COLOR_BGR2RGB))
axs[0, 1].set_title('Reddish Image')
axs[0, 1].axis('off')

# Greenish Image
axs[0, 2].imshow(cv2.cvtColor(img_greenish, cv2.COLOR_BGR2RGB))
axs[0, 2].set_title('Greenish Image')
axs[0, 2].axis('off')

# Bluish Image
axs[1, 0].imshow(cv2.cvtColor(img_bluish, cv2.COLOR_BGR2RGB))
axs[1, 0].set_title('Bluish Image')
axs[1, 0].axis('off')

# Yellowish Image
axs[1, 1].imshow(cv2.cvtColor(img_yellowish, cv2.COLOR_BGR2RGB))
axs[1, 1].set_title('Yellowish Image')
axs[1, 1].axis('off')

# Hide the last subplot (bottom right)
axs[1, 2].axis('off')   # Hide the last subplot

# Display the plot
plt.show()
```

> ## Eliminating color channels >> 

> 

```python
# Eliminate the red channel
img_no_red = img.copy()
img_no_red[:, :, 2] = 0  # Set the red channel to 0

# Eliminate the green channel
img_no_green = img.copy()
img_no_green[:, :, 1] = 0  # Set the green channel to 0

# Eliminate the blue channel
img_no_blue = img.copy()
img_no_blue[:, :, 0] = 0  # Set the blue channel to 0

# Display the images
cv2.imshow('Original Image', img)
cv2.imshow('No Red Channel', img_no_red)
cv2.imshow('No Green Channel', img_no_green)
cv2.imshow('No Blue Channel', img_no_blue)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
