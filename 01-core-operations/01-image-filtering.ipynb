````python
import cv2
import numpy as np
````

> ## ```filter2D(src,ddepth,kernel)```
>

````python
# The depth of an image refers to the number of bits used to represent each pixel

image = cv2.imread('apple1.png')

# Create a kernel
kernel = np.array(  [[0, -1, 0], 
                    [-1, 5,-1], 
                    [0, -1, 0]] )

# Apply the filter
filtered_image = cv2.filter2D(src=image, ddepth=-1, kernel=kernel) 

cv2.imshow('Original Image', image)
cv2.imshow('Filtered Image', filtered_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````

> ## blur(src,size) & medianBlur(src,size) 
> 

````python
# Apply a mean blur "normal"
blurred_image = cv2.blur(image, (5, 5))

# Apply a median blur
median_blurred_image = cv2.medianBlur(image, 5)

cv2.imshow("original", image)
cv2.imshow('mean Blurred Image', blurred_image)
cv2.imshow('Median Blurred Image', median_blurred_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
````
