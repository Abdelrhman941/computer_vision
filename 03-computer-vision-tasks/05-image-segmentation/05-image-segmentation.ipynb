```python
image1 = cv2.imread('photo1.jpeg')

## to convert the image in grayscale  
img = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

## ret is the threshold value that we have used,
ret, thresh1 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)  
cv2.imshow('Binary Threshold', thresh1) 
print(ret)

ret, thresh2 = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY_INV) 
cv2.imshow('Binary Threshold Inverted', thresh2) 
print(ret)

ret, thresh3 = cv2.threshold(img, 120, 255, cv2.THRESH_TRUNC) 
cv2.imshow('Truncated Threshold', thresh3) 
print(ret)

ret, thresh4 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO) 
cv2.imshow('Set to 0', thresh4) 
print(ret)

ret, thresh5 = cv2.threshold(img, 120, 255, cv2.THRESH_TOZERO_INV) 
cv2.imshow('Set to 0 Inverted', thresh5)
print(ret)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## Displaying the images by using matplotlib >>
titles = ['Original Image', 'Binary Threshold', 'Binary Threshold Inverted', 'Truncated Threshold', 'Set to 0', 'Set to 0 Inverted']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

plt.figure(figsize=(10, 8)) 

for i in range(6):
    plt.subplot(2, 3, i+1)  # 2 rows, 3 columns , i+1 is the index
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])  # to hide the x and y axis

plt.show()
```

> ## Contour Detection
> 

```python
img = cv2.imread('shapes-small.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('Original image')
plt.axis('off')  # Hide the axis
plt.show()
```

```python
imgray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, binary = cv2.threshold(imgray, 100, 255, cv2.THRESH_OTSU)
inverted_binary = ~binary   # ~ is the bitwise_not operator


cv2.imshow('Gray image', imgray)
cv2.imshow('Binary image', binary)
cv2.imshow('Inverted binary image', inverted_binary)

cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
contours, hierarchy = cv2.findContours(inverted_binary , cv2.RETR_TREE , cv2.CHAIN_APPROX_SIMPLE)   # findContours() function to detect the contours in the image
## contours  >> returns a list of contours 
## hierarchy >> returns the parent-child relationship of contours

with_contours = cv2.drawContours(img, contours, -1,(255,0,255),3)
## Draw the contours (in red) on the original image and display the result
## Input color code is in BGR (blue, green, red) format
## -1 means to draw all contours

cv2.imshow('Original image', img)
cv2.imshow('Detected contours', with_contours)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
